```python
#Lấy dữ liệu ở trong file csv
f = open('new_data.csv', 'r',encoding='utf8')
line = f.readline()
toan = []
van = []
anh =  []
ly = []
hoa = []
sinh = []
su = []
dia = []
gdcd = []
for line in f.readlines():
    line = line.split(',')
    if line[3] != '0' :
        toan.append(float(line[3]))
    if line[4] != '0' :
        van.append(float(line[4]))
    if line[5] != '0' :
        anh.append(float(line[5]))
    if line[6] != '0' :
        ly.append(float(line[6]))
    if line[7] != '0' :
        hoa.append(float(line[7]))
    if line[8] != '0' :
        sinh.append(float(line[8]))
    if line[9] != '0' :
        su.append(float(line[9]))
    if line[10] != '0' :
        dia.append(float(line[10]))
    if line[11][0] !='0':
        gdcd.append(float(line[11]))
f.close()
# Close file
```

```python
import numpy as np
toan = np.array(toan)
van = np.array(van)
anh = np.array(anh)
ly = np.array(ly)
hoa = np.array(hoa)
sinh = np.array(sinh)
su = np.array(su)
dia = np.array(dia)
gdcd = np.array(gdcd)
```

```python
# Môn toán 
from matplotlib import pyplot as plt
%matplotlib inline 

# Trung bình và trung vị 
toan.mean(), np.median(toan)

# Xem có bao nhiêu đầu điểm
unique = np.unique(toan)
print(unique)

# Đếm xem mỗi đầu điểm có bao nhiêu người
a = list(toan)
freq = [a.count(b) for b in unique]
print(freq)

# Đồ thị
plt.bar(unique, freq, width=0.3)
plt.xticks(unique, rotation=90)
plt.xlabel(" Đầu điểm môn Toán ")
plt.ylabel(" Số lượng học sinh ")
```
